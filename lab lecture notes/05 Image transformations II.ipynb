## Rotation of Image ##

The rotation matrix is given by



(center.x, center.y) is the center of rotation 


```python
# get h,w of original image
img = cv2.imread('C:/Users/USER/Desktop/AI Lab/Data/01-Data/lena.png')
h, w = img.shape[:2]
```

```python
# define rotation matrix
M = cv2.getRotationMatrix2D((h//2,w//2), 45, 1) #(center, angle, scale factor)
```

```python
# apply rotation to original image
new_img = cv2.warpAffine(img, M, (h,w)) # (image, matrix, dimensions)
```

```python
cv2.imshow("Rotated Image", new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### PROBLEM: Given a final image and an initial one, how to find the transformation matrix that led the image there?
Practical examples:
- Panorama image creation (how to glue images to make the panorama)
- Low level feature extraction 

```python
# create 2 sets of points
pts_1 = np.float32([
    [135,45],
    [385,45],
    [135,230]
])

pts_2 = np.float32([
    [135,45],
    [385,45],
    [150,230]
])
```

```python
# get the matrix M that transforms pts_1 to pts_2
M = cv2.getAffineTransform(pts_1, pts_2) #you cannot always find affine transformation!
new_img = cv2.warpAffine(pts_1, M, (1024,1024))
print(M)
```
